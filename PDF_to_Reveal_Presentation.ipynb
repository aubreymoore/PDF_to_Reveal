# Step 2: Convert PDF pages to PNG images

We can do this using the **convert** command from Imagemagick. The conversion will take a few minutes. When finished the **jpg** directory will be populated with files named 000.jpg, 001.jpg, etc. 

# Step 3: Insert JPG images into reveal.js HTML template

```python
jpg_list = sorted(os.listdir('jpg'))

first_slide_md = '''
# Note

Images in this **reveal.js** slide deck were extracted using this ImageMagick command:

    convert -density 150 -antialias BioinvasionOfGuamForests.pdf
    -resize 1024x -quality 100 jpg/%03d.jpg
    
and inserted into an HTML template using the method detiled in my Jupyter notebook 
entitled **PDF_to_Reveal_Presentation**.  
Available from <https://github.com/aubreymoore/PDF_to_Reveal/blob/master/PDF_to_Reveal.ipynb>

Aubrey Moore

University of Guam
'''

mydict = {'first_slide_md': first_slide_md, 'jpg_list': jpg_list}
```

```python
env = Environment(loader=FileSystemLoader('templates'))
template = env.get_template('reveal.html')
output_from_parsed_template = template.render(mydict)
print(output_from_parsed_template)

# to save the results
with open("reveal.js/slides.html", "w") as f:
    f.write(output_from_parsed_template)
```
